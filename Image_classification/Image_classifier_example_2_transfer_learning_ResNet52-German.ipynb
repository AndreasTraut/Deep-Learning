Zu Beginn müssen einige Bibliotheken importiert werden. Das ist ganz einfach: 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.applications.imagenet_utils import decode_predictions
```

Zunächst lade ich das Bild: 

```python
fig, ax = plt.subplots(1, figsize=(12, 10))
img = image.load_img('../../Deep_learning_explorations/data/caltechown/102/9145.JPG', target_size=(224, 224))
img = image.img_to_array(img)
ax.imshow(img / 255.) 
ax.axis('off')
plt.show()
```

Im Folgenden lade ich das "ResNet50" Modell, welches bereits auf dem "ImageNet"-Datensatz vortrainiert wurde und übergebe das obige Bild der "predict" Funktion. Diese Funktion gibt mir dann eine Aussage zurück, was das Modell auf dem Bild sieht. In meinem Fall sieht es eine Tasse ("Cup") mit 86% Wahrscheinlichkeit oder mit 6.7% Wahrscheinlichkeit ein "coffeepot": 

```python
resnet = ResNet50(weights='imagenet')
x = preprocess_input(np.expand_dims(img.copy(), axis=0))
preds = resnet.predict(x)
decode_predictions(preds, top=5)
```
